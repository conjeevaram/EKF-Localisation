```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def generate_trajectory(num_points=100):
    # Create a spiral in the x-y plane that progresses along the z axis.
    # z increases steadily, while the spiral radius in the x-y plane shrinks gently as z increases.
    t = np.linspace(0, 4 * np.pi, num_points)  # angular parameter
    z = np.linspace(0, 10, num_points)         # z-axis progression
    # Radius decreases gently: from 1 to 0.7
    radius = np.linspace(1, 0.7, num_points)
    x = radius * np.cos(t)
    y = radius * np.sin(t)
    return x, y, z

def plot_trajectory(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot3D(x, y, z, 'blue')
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    ax.set_title("Trajectory")
    plt.show()

# Generate and plot the rotated spiral trajectory
x, y, z = generate_trajectory()
plot_trajectory(x, y, z)
```

```python
# Generate more points by passing a larger num_points value
x, y, z = generate_trajectory(num_points=1000)

# Then apply Gaussian noise
noise_std = 0.1  # adjust the noise std as needed
noisy_x = x + np.random.normal(0, noise_std, x.shape)
noisy_y = y + np.random.normal(0, noise_std, y.shape)
noisy_z = z + np.random.normal(0, noise_std, z.shape)

# Plot the noisy trajectory as a point cloud
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(noisy_x, noisy_y, noisy_z, color='red', s=20, label='Noisy Trajectory')

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Noisy Trajectory")
ax.legend()
plt.show()
```
